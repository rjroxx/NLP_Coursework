```
url="http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz"
```

```
df=pd.read_json(url,lines=True)
dp=df.head(50)
```

```
def custom_func(text):
  
  wt=word_tokenize(text)
  counter = {'tokenizer': wt} 
  return counter
```

```
dp['reviewText_nltk_tokenizer'] = dp.reviewText.apply(custom_func)
```

```
def custo(text):
  
  wt=word_tokenize(text)
  pt=nltk.pos_tag(wt)
  counter = {'Pos_tag': pt} 
  return counter
```

```
dp['reviewText_nltk_Pos_tag'] = dp.reviewText.apply(custo)
```

```
def customo(text):
  
  wt=word_tokenize(text)
  pt=dict(nltk.pos_tag(wt))
  ptr=[word for word in pt if pt[word]=='JJ']
  counter = {'Adjective': ptr} 
  return ptr
```
