```
import pandas as pd
from collections import Counter
import string
import re
```

```
url="http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz"
```

```
df=pd.read_json(url,lines=True)
```

Vowel,Consonant & Word Count

```
def custom_func(text):
  vc=0
  cc=0
  for i in text:
      if( i=='A' or i=='a' or i=='E' or i=='e' or i=='I'
          or i=='i' or i=='O' or i=='o'
        	or i=='U' or i=='u'):
          	vc +=1
      elif(i=='B' or i=='b' or i=='c' or i=='C' or i=='D'
          or i=='d' or i=='F' or i=='f' or i=='G' or i=='g' 
          or i=='H' or i=='h' or i=='J' or i=='j' or i=='K'
          or i=='K' or i=='L' or i=='l'	or i=='M' or i=='m' 
          or i=='N' or i=='n' or i=='P' or i=='p' or i=='Q'
          or i=='q' or i=='R' or i=='r' or i=='S' or i=='s' 
          or i=='T' or i=='t' or i=='V' or i=='v' or i=='W'
          or i=='w' or i=='X' or i=='x' or i=='Y' or i=='y' or i=='Z' or i=='z'):
           cc +=1 
      else:
        continue
  words = text.split()
  wordCount = len(words)
  counter = {'vowel': vc,'con': cc,'WordCount':wordCount} 
  return counter
```

```
df['reviewTextCount'] = df.reviewText.apply(custom_func)
```

Words that Starts with B and Ends with G


```
def custom_func(text):
  words = re.findall(r'\b[bB]\S+[gG]\b', text)
  return words
```

```
df['reviewTextCount'] = df.reviewText.apply(custom_func)
```
